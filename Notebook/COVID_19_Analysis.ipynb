### Exploratory Data Analysis (EDA)

```python

# state/country data producer
def stateTS(df, stateName, getGraph=False, newCol=False):
  if stateName == 'India':
    indiaDF = pd.DataFrame(df.sum())
    indiaDF = indiaDF.reset_index().iloc[1:]
    indiaDF.columns = ['date', 'India']
    # set new df for country
    transp_thisDF = indiaDF.copy()
  else:
    thisDF = df[:][df.State == stateName]
    # trnsposeing
    transp_thisDF = thisDF.T
    transp_thisDF = transp_thisDF.reset_index().iloc[1:]
    transp_thisDF.columns = ['date', stateName]

  # adding new col (common code)
  cumulativeCases = list(transp_thisDF[stateName])
  newCases = []
  for i in range(len(cumulativeCases)-1,-1,-1):
    if i > 0: 
      newCaseCount = cumulativeCases[i] - cumulativeCases[i-1]
    else:
      newCaseCount = cumulativeCases[i]
    newCases.append(newCaseCount)
  # renaming cols
  transp_thisDF['newCol'] = newCases[::-1]
  # rem first row        
  transp_thisDF.drop([1], axis=0,inplace=True)
  transp_thisDF = transp_thisDF.reset_index(drop=True)
  transp_thisDF = transp_thisDF.reset_index()
  transp_thisDF.columns = ['seq', 'date', 'cumulative', 'new']
  # req for model
  transp_thisDF.seq += 1

  if getGraph:
    if newCol:
      fig = px.line(transp_thisDF, x="date", y=["new"], template = 'simple_white', title="Graph for "+stateName)
    else:
      fig = px.line(transp_thisDF, x="date", y=["cumulative"], template = 'simple_white', title="Graph for "+stateName)
    fig.show()
  
  return transp_thisDF
```
